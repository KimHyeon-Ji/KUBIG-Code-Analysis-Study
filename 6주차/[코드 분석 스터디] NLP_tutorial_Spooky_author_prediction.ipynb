## NLP 

우리는 소설의 작가를 예측하기 위해서는 NLP를 진행하여야한다. 위에 있는 그림들은 단순히 그 작가의 소설에서 많이 나오는 단어들을 시각화시켜준 것이며 이는 예측이라기 보단 EDA의 과정이다. 컴퓨터는 textual data에 대한 처리를 하지 못하기 때문에 사용자는 학습 전에 가공을 해야한다. 

### Tokenization
 문자열들을 작은 단위로 자른다. 이 커널에서는 내제되어 있는 라이브러리를 사용할 것이다.
 
### Stopwords
 예측력이 떨어지는 단어의 경우는 배제한다. 이에 대한 구분은 빈도를 표현하기도 한다.
 
### Stemming
 단어 속에서 뜻을 담는 곳을 추출하는 과정이다. ex) Sleep, Sleepy, Sleeping -> sleep
 'root'-'Stem'으로 나눈다. 형태소를 제거한다
 
### Vectorization 
 NLP를 컴퓨터가 이해할 수 있도록 수치로 바꾸는 것을 의미한다. 이 때 벡터로 변환된 고유의 토큰들이 모인 집합을 vocabulary 하며 vocabulary가 크면 클수록 학습에 오래 걸린다.

```python
lemm = WordNetLemmatizer()
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc : (lemm.lemmatize(w) for w in analyzer(doc))
```

```python
#Storing the entire training text in a list
text = list(train.text.values)
#Calling our overwritten Count vectorizer
tf_vectorizer = LemmaCountVectorizer(max_df=0.95, min_df=2,stop_words='english',decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
#Complete Dictionary of our words
print(tf_vectorizer.get_feature_names())
```

```python
feature_names = tf_vectorizer.get_feature_names()
count_vec = np.asarray(tf.sum(axis=0)).ravel()
zipped = list(zip(feature_names, count_vec))
x, y = (list(x) for x in zip(*sorted(zipped, key=lambda x: x[1], reverse=True)))
#Now I want to extract out on the top 15 and bottom 15 words
Y = np.concatenate([y[0:15], y[-16:-1]])
X = np.concatenate([x[0:15], x[-16:-1]])

#Plotting the Plot. ly plot for the Top 50 word frequencies
data = [go.Bar(
    x=x[0:50], y=y[0:50], marker = dict(colorscale='Viridis', color = y[0:50]), text ='Word counts')]
layout = go.Layout(
title='Top 50 word frequencies after Preprocessing')
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='basic-bar')
data = [go.Bar(x=x[-100:],
              y=y[-100:],
              marker = dict(colorscale='Jet', color = y[-100:]),
              text = 'Word counts')]
layout = go.Layout(title='Bottom 100 Word frequencies after Preprocessing')
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='basic-bar')
```

```python
lda = LatentDirichletAllocation(n_components=11, max_iter=5, learning_method='online',learning_offset=50, random_state=0)
lda.fit(tf)
lda.components_
```

```python
#Define helper function to print top words
def print_top_words(model, feature_names, n_top_words):
    for index, topic in enumerate(model.components_):
        message = '\nTopic #{}:'.format(index)
        message += ' '.join([feature_names[i] for i in topic.argsort()[:-n_top_words-1:-1]])
        print(message)
        print('='*70)
```

```python
number_of_words=50
print('\nTopic in LDA model: ')
tf_feature_names=tf_vectorizer.get_feature_names()
# note that tf_vectorizer is an LemmaCountVectorizer object and with this command we get the whole dictionary of words
print_top_words(lda, tf_feature_names, number_of_words)
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
```

```python
print(first_topic)
```

```python
first_topic_words = [tf_feature_names[i] for i in first_topic.argsort()[:-50 - 1 :-1]]
second_topic_words = [tf_feature_names[i] for i in second_topic.argsort()[:-50 - 1 :-1]]
third_topic_words = [tf_feature_names[i] for i in third_topic.argsort()[:-50 - 1 :-1]]
fourth_topic_words = [tf_feature_names[i] for i in fourth_topic.argsort()[:-50 - 1 :-1]]
```
